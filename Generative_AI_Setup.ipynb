```python
from diffusers import StableDiffusionPipeline

# Load the Stable Diffusion model from Hugging Face
pipeline = StableDiffusionPipeline.from_pretrained("runwayml/stable-diffusion-v1-5", torch_dtype=torch.float16)
pipeline = pipeline.to("cuda")  # Move the model to GPU
```

```python
prompt = "A beautiful landscape with mountains and a lake during sunset"
image = pipeline(prompt).images[0]

```

```python
import os
from google.cloud import storage

# Define the local directory where the dataset was extracted
local_directory = '/content/flickr30k'

# Initialize the Google Cloud Storage client and get the bucket
bucket_name = 'text-to-image-dataset-bucket-assignment-2-443109'  # Replace with your actual bucket name
client = storage.Client()
bucket = client.get_bucket(bucket_name)

# Function to upload files recursively
def upload_files(local_directory, bucket):
    for root, dirs, files in os.walk(local_directory):  # Walk through subdirectories
        for filename in files:
            local_file = os.path.join(root, filename)
            # Construct the path in the cloud storage to maintain directory structure
            cloud_path = os.path.relpath(local_file, local_directory)
            blob = bucket.blob(cloud_path)
            blob.upload_from_filename(local_file)
            print(f'Uploaded {filename} to {bucket_name}/{cloud_path}')

# Upload the dataset
upload_files(local_directory, bucket)
```
